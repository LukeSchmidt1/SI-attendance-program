```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/lukeschmidt/Downloads/MENCE Attendance Forms - Fall 2024(Observations).csv')
df
```

```python
df.drop(columns=['Observation 1', 'Observation 2', 'Observation 3 '], inplace=True)
df
```

```python
df[['first_name', 'last_name']] = df['SI/Pass+ Leaders'].str.split(n=1, expand=True)
df.drop(columns=['SI/Pass+ Leaders'], inplace=True)

df
```

```python
df.drop(columns=['Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10'], inplace=True)
df
```

```python
# Split by space to separate day(s) from time
df[['days', 'time']] = df['Session Days & Times'].str.split(' ', n=1, expand=True)

# Further split days if multiple days are listed (like Monday/Wednesday)
df[['day1', 'day2']] = df['days'].str.split('/', n=1, expand=True)

# Optionally, drop the old 'Session Days & Times' and 'days' columns
df.drop(columns=['Session Days & Times', 'days'], inplace=True)
df
```

```python
from sqlalchemy import create_engine

# Database configuration
db_config = {
    'user': 'root',           # Your MySQL username
    'password': '',           # Your MySQL password
    'host': 'localhost',
    'database': 'leaders_db',      # Starting with the test database
    'raise_on_warnings': True,
}

# Create the SQLAlchemy engine
engine = create_engine(f'mysql+pymysql://{db_config["user"]}:{db_config["password"]}@{db_config["host"]}/{db_config["database"]}')

# Insert the DataFrame into the MySQL table
df.to_sql('leaders', con=engine, if_exists='append', index=False)

# Close the engine connection (optional)
engine.dispose()
```
